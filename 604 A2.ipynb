```python
# correlation between total sales by country and product line
sales_by_country_productline = df.groupby(['COUNTRY', 'PRODUCTLINE'])['SALES'].sum().unstack()
plt.figure(figsize=(15, 10))
sns.heatmap(sales_by_country_productline, annot=True, fmt='.2f', cmap='Blues')
plt.title('Total Sales by Country and Product Line')
plt.xlabel('Product Line')
plt.ylabel('Country')
plt.show()
```

```python
# calculate the mean of priceeach and msrp for each product
avg_prices = df.groupby('PRODUCTLINE')[['PRICEEACH', 'MSRP']].mean().reset_index()

# set the plot size
bar_width = 0.25
x = range(len(avg_prices))
plt.figure(figsize=(10, 8))

# plot PRICEEACH and MSRP
plt.bar(x, avg_prices['PRICEEACH'], width=bar_width, label='PRICEEACH', color='b', alpha=0.6)
plt.bar([i + bar_width for i in x], avg_prices['MSRP'], width=bar_width, label='MSRP', color='r', alpha=0.6)

# add labels and title
plt.xlabel('Product')
plt.ylabel('Price')
plt.title('Comparison of PRICEEACH and MSRP by Product')
plt.xticks([i + bar_width / 2 for i in x], avg_prices['PRODUCTLINE'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Group the data by YEAR_ID and STATUS and count the number of orders for each group
status_by_year = df.groupby(['YEAR_ID', 'STATUS']).size().reset_index(name='count')

# Get unique years
years = status_by_year['YEAR_ID'].unique()

# Set up the subplot layout
fig, axes = plt.subplots(1, len(years), figsize=(20, 8))

# Create a pie chart for each year
for i, year in enumerate(years):
    year_data = status_by_year[status_by_year['YEAR_ID'] == year]
    axes[i].pie(year_data['count'], labels=year_data['STATUS'], autopct='%1.1f%%', startangle=140,pctdistance=1.1, labeldistance=1.2)
    axes[i].set_title(f'Order Status Distribution in {year}')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
from scipy.stats import pearsonr
import statsmodels.formula.api as sm
import statsmodels.api as sm_api
from scipy.stats import chi2_contingency
from statsmodels.formula.api import ols
```

```python
# generate a list of product lines and oerform the one-way ANOVA
product_lines = ['Motorcycles','Classic Cars','Trucks and Buses','Vintage Cars','Planes','Ships','Trains']
anova_model = ols('SALES ~ C(PRODUCTLINE)', data=df).fit()
anova_table = sm_api.stats.anova_lm(anova_model, typ=2)
print(anova_table)
```

```python
# perform Pearson correlation test
correlation, p_value = pearsonr(df['QUANTITYORDERED'], df['PRICEEACH'])
print(f"Pearson Correlation: {correlation}, p-value: {p_value}")
```

```python
# perform Chi-square test
contingency_table = pd.crosstab(df['DEALSIZE'], df['STATUS'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square statistic: {chi2}")
print(f"P-value: {p}")
```
